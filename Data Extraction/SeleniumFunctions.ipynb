## Get driver

```python
def getdriver():
  from selenium import webdriver
  from selenium.webdriver.chrome.options import Options
  import uuid
  options = Options()
  options.add_argument("--headless")
  options.add_argument("--no-sandbox")
  options.add_argument("--disable-dev-shm-usage")
  # Set a unique user data directory
  options.add_argument(f"--user-data-dir=/tmp/chrome-profile{str(uuid.uuid4())}")
  driver = webdriver.Chrome(options=options)
  return driver

driver = getdriver()
```

## Get Soup

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup

wait = WebDriverWait(driver, 5)

def getsoup(url, by_id = None, by_xpath = None, by_linktext = None, by_partiallinktext = None, by_name = None, by_tag = None, by_class = None, by_css = None):
  driver.get(url)

  if by_id:
    wait.until(EC.presence_of_element_located((By.ID, by_id)))
  elif by_xpath:
    wait.until(EC.presence_of_element_located((By.XPATH, by_xpath)))
  elif by_linktext:
    wait.until(EC.presence_of_element_located((By.LINK_TEXT, by_linktext)))
  elif by_partiallinktext:
    wait.until(EC.presence_of_element_located((By.PARTIAL_LINK_TEXT, by_partiallinktext)))
  elif by_name:
    wait.until(EC.presence_of_element_located((By.NAME, by_name)))
  elif by_tag:
    wait.until(EC.presence_of_element_located((By.TAG_NAME, by_tag)))
  elif by_class:
    wait.until(EC.presence_of_element_located((By.CLASS_NAME, by_class)))
  elif by_css:
    wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, by_css)))
  else:
    wait.until(EC.presence_of_element_located((By.TAG_NAME, "body")))

  html = driver.page_source
  soup = BeautifulSoup(html, "html.parser")
  return soup

BASE_URL = 'https://jivraj-18.github.io/tds-jan-2025-mock-roe-1/'
soup = getsoup(BASE_URL)
print(soup.prettify())
```
